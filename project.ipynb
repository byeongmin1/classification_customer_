```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as  sns
```

시각화

```python
#normalization
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(train.drop(['depvar','purpose4'], axis=1)), columns=train.drop(['depvar','purpose4'], axis=1).columns)
```

```python
X.columns
```

```python
y = train['depvar']
```

```python
print((X.shape,y.shape))
```

```python
# train valid split
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import sklearn.metrics as metrics
```

```python
X_train, X_valid, y_train, y_valid =train_test_split(X,y, test_size=0.25, random_state=42)
```

```python
print(f'training shape: {X_train.shape}, {y_train.shape}')
print(f'valid shape: {X_valid.shape}, {y_valid.shape}')
```

```python
y_valid.sum()/len(y_valid)
```

```python
y_train.head()
```

```python

lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
lr.predict_proba(X_valid)
```

```python
y_pred = (lr.predict_proba(X_valid)[:, 1] >= 0.27).astype(int)
```

```python
# y_pred = lr.predict(X_valid)
```

```python
def get_clf_eval(y_answer, y_pred):
    acc = metrics.accuracy_score(y_answer,y_pred)
    prec = metrics.precision_score(y_answer, y_pred)
    recall = metrics.recall_score(y_answer,y_pred)
    AUC = metrics.roc_auc_score(y_answer, y_pred)
    F1 = metrics.f1_score(y_answer,y_pred)

    print("정확도: {:.6f}".format(acc))
    print("정밀도: {:.6f}".format(prec))
    print("재현율: {:.6f}".format(recall))
    print("AUC: {:.6f}".format(AUC))
    print("F1: {:.6f}".format(F1))


    sns.heatmap(metrics.confusion_matrix(y_answer, y_pred), annot= True, fmt='d', cmap='YlGnBu')
    plt.xlabel("Prediction labels")
    plt.ylabel("True albels")
    plt.show()
```

```python
get_clf_eval(y_valid,y_pred)
```
